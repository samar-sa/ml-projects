```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Linear Regression with Synthetic data**

```python
from numpy.random import randn
X = randn(50)
print(X.shape)
alpha = 2.5
beta = 4.0
Y = alpha*X + beta
```

```python
plt.rcParams['figure.figsize'] = (6.0, 4.0)
plt.scatter(X, Y)
plt.show()
```

**Building Linear regaression**


```python
X_mean = np.mean(X)
Y_mean = np.mean(Y)
num = 0
den = 0
for i in range(len(X)):
 num += (X[i] - X_mean)*(Y[i] - Y_mean)
 den += (X[i] - X_mean)**2
w_opt = num / den
b_opt = Y_mean - alpha*X_mean
print (alpha, beta)
print (w_opt,b_opt)
```

**Prediction and Computing the Sum Square Error**

```python
Y_pred= w_opt*X+ b_opt
SSE = np.sum(np.square(Y-Y_pred))
print("Sum Square Error is equal to:", SSE)
```

**Compute R2**

```python
SS_tot = np.sum(np.square(Y - np.mean(Y)))
SS_res = SSE
R2 = 1- (SS_res/SS_tot)
print ("R2 is equal to:", R2)
```

```python
plt.rcParams['figure.figsize'] = (6.0, 4.0)
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.show()
```
